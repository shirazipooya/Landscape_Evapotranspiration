```python
Coeff = Kd.merge(
    right=Kmc, how="outer", on=["نوع لکه", "کد نوع لکه"]
).merge(
    right=Ks, how="outer", on=["نوع گونه", "کد نوع گونه", "ماه"]
).merge(
    right=IE, how="outer", on=["نوع گونه", "کد نوع گونه", "نوع لکه", "کد نوع لکه"]
)

Coeff = Coeff[["نوع لکه",
               "کد نوع لکه",
               "نوع گونه",
               "کد نوع گونه",
               "روش آبیاری",
               "کد روش آبیاری",
               "ماه",
               "Ks",
               "Kd",
               "Kmc",
               "IE"]]

Coeff.columns = ["نام نوع لکه",
                 "نوع لکه",
                 "نام نوع قلم",
                 "نوع قلم",
                 "نام زیرمجموعه هر قلم",
                 "زیرمجموعه هر قلم",
                 "ماه",
                 "Ks",
                 "Kd",
                 "Kmc",
                 "IE"]

Coeff = Coeff.pivot(index=["نام نوع لکه", "نوع لکه", "نام نوع قلم", "نوع قلم", "نام زیرمجموعه هر قلم", "زیرمجموعه هر قلم"], columns='ماه', values=["Ks", "Kd", "Kmc", "IE"])\
.reset_index()

Coeff.columns = ['_'.join(col).strip() for col in Coeff.columns.values]

Coeff.rename(columns={"نام نوع لکه_" : "نام نوع لکه",
                     "نوع لکه_" : "نوع لکه",
                     "نام نوع قلم_" : "نام نوع قلم",
                     "نوع قلم_" : "نوع قلم",
                     "نام زیرمجموعه هر قلم_" : "نام زیرمجموعه هر قلم",
                     "زیرمجموعه هر قلم_" : "زیرمجموعه هر قلم"},
             inplace=True)
```

        <h2>اضافه کردن ضرایب به دیتابیس</h2>
    </font>

```python
Kd_v2 = {
    1: 1,
    2: 1,
    3: 1.2,
    4: 1.2,
    5: 1.3,
    6: 1.3,
    7: 1.3,
    8: 1.3
}
```

```python
IDs_Asle = ["08", "09"]

dataset_coeff = pd.DataFrame()

for idf, df in dataset.groupby(["منطقه", "ناحیه", "پیمان", "آدرس"]):
    
    if df.query('`زیرمجموعه هر قلم` != @IDs_Asle').shape[0] != 0:        
    
        tmp = df.query('`زیرمجموعه هر قلم` != @IDs_Asle')\
        .groupby(["منطقه", "ناحیه", "پیمان", "آدرس", "نوع قلم"])\
        .apply(lambda x : x['مقدار آیتم'].sum())\
        .reset_index()\
        .rename(columns={0 : "مساحت گونه"})\
        .assign(**{"درصد مساحت گونه در لکه" : lambda x : x['مساحت گونه'] / x['مساحت گونه'].sum()})

        df = df.merge(tmp, how='left', on=["منطقه", "ناحیه", "پیمان", "آدرس", "نوع قلم"])

        df.loc[df["زیرمجموعه هر قلم"] == "08", ["مساحت گونه", "درصد مساحت گونه در لکه"]] = 0
        df.loc[df["زیرمجموعه هر قلم"] == "09", ["مساحت گونه", "درصد مساحت گونه در لکه"]] = 0

        Number_Species = tmp["نوع قلم"].unique().__len__()

        if "08" in list(df["زیرمجموعه هر قلم"]):
            Number_Species += 1

        df["تعداد گونه در لکه"] = Number_Species

        df['Kd_v2'] = list(map(Kd_v2.get, df['تعداد گونه در لکه']))

        df = df.merge(Coeff.drop(["نام نوع لکه", "نام نوع قلم", "نام زیرمجموعه هر قلم"], axis=1),
                      how="left",
                      on=["نوع لکه", "نوع قلم", "زیرمجموعه هر قلم"])

        dataset_coeff = pd.concat([dataset_coeff, df])
    else:
        df = df.assign(**{"مساحت گونه" : 0})
        df = df.assign(**{"درصد مساحت گونه در لکه" : 0})
        df = df.assign(**{"تعداد گونه در لکه" : 1})
        df = df.assign(**{"Kd_v2" : list(map(Kd_v2.get, df['تعداد گونه در لکه']))})
        
        df = df.merge(Coeff.drop(["نام نوع لکه", "نام نوع قلم", "نام زیرمجموعه هر قلم"], axis=1),
                      how="left",
                      on=["نوع لکه", "نوع قلم", "زیرمجموعه هر قلم"])
        
        dataset_coeff = pd.concat([dataset_coeff, df])

dataset_coeff = dataset_coeff.reset_index(drop=True)
```

<div style="direction:rtl;line-height:300%;" id="section3">
    <font size=3>
        <center>$LET = PET * K_s * K_d * K_{mc}$</center>
        <center>$IR = \frac{LET - Eff.Precip}{IE} * Area$</center> 
    </font>

<div style="direction:rtl;line-height:150%;">
    <font face="B Zar">
        <h3>روش اول (حمید عمرانیان)</h3>
    </font>  
</div>

```python
Month_Name = ["M01", "M02", "M03", "M04", "M05", "M06", "M07", "M08", "M09", "M10", "M11", "M12"]

result_method_01 = dataset_coeff.copy()

for i, value in enumerate(Month_Name):    
    result_method_01["KL_" + value] = result_method_01["Ks_" + value] * result_method_01["Kd_" + value] * result_method_01["Kmc_" + value]

for i, value in enumerate(Month_Name):    
    result_method_01["LET_" + value] = result_method_01["PET_" + value] * result_method_01["KL_" + value]

for i, value in enumerate(Month_Name):
    result_method_01["IR_" + value] = ((result_method_01["LET_" + value] - result_method_01["Effe_Precip_" + value]) * result_method_01["n_Day_" + value] / result_method_01["IE_" + value]) * result_method_01["مقدار آیتم"] / 1000
    result_method_01.loc[result_method_01["IR_" + value] < 0, ["IR_" + value]] = 0

result_method_01["IR"] = result_method_01[["IR_M01", "IR_M02", "IR_M03", "IR_M04", "IR_M05", "IR_M06", "IR_M07", "IR_M08", "IR_M09", "IR_M10", "IR_M11", "IR_M12"]].sum(axis=1)

file_name = PROJECT_PATH + "assets/report/Method_01.xlsx"
with pd.ExcelWriter(path=file_name, engine="openpyxl", mode="w") as writer:
    result_method_01.to_excel(writer, index=True, sheet_name="نتایج کلی")
```
